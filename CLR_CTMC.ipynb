```python
import sympy as sp

# 1. Define the symbols
omega, nu, U_star, S_star, mu, d, gamma, mu_U, delta, eta, mu_P, lam = sp.symbols(
    'omega nu U^* S^* mu d gamma mu_U delta eta mu_P lambda'
)

# 2. Define the 3x3 submatrix M
M = sp.Matrix([
    [-omega*nu*U_star - mu, 0, -omega*nu*S_star],
    [omega*nu*U_star, -(mu+d), omega*nu*S_star],
    [-nu*U_star, gamma, -nu*S_star - mu_U]
])

# 3. Compute Characteristic Polynomial Coefficients
# Returns coeffs for: lambda^3 + a1*lambda^2 + a2*lambda + a3
char_poly = M.charpoly(lam)
coeffs = char_poly.all_coeffs()

a3 = sp.simplify(coeffs[0])
a2 = sp.simplify(coeffs[1])
a1 = sp.simplify(coeffs[2])
a0 = sp.simplify(coeffs[3])

# 4. Compute Routh-Hurwitz condition: a1*a2 - a3
rh_condition = sp.simplify(a2 * a1 - a3*a0)
```

### Summary of Routh-Hurwitz Stability Conditions

For a system with a characteristic polynomial $P(\lambda) = \lambda^3 + a_2\lambda^2 + a_1\lambda + a_0$, stability according to the Routh-Hurwitz criterion requires the following conditions to be met:

1.  **All coefficients must be positive**:
    *   $a_2 > 0$
    *   $a_1 > 0$
    *   $a_0 > 0$

2.  **Routh-Hurwitz determinant must be positive**:
    *   $a_1 a_2 - a_0 > 0$

These conditions ensure that all roots of the characteristic polynomial have negative real parts, which is necessary for the stability of the system at the equilibrium point.

```python
import sympy as sp

# ==========================================
# 1. DÉFINITION DES VARIABLES SYMBOLIQUES
# ==========================================
print("--- 1. INITIALISATION DES VARIABLES ---")
lam = sp.symbols('lambda')
omega, nu, mu, d, gamma, mu_U, delta, eta, mu_P = sp.symbols(
    'omega nu mu d gamma mu_U delta eta mu_P', real=True, positive=True
)
S, I, U, P = sp.symbols('S_star I_star U_star P_star', real=True, positive=True)

# ==========================================
# 2. DÉFINITION DE LA MATRICE JACOBIENNE J
# ==========================================
# Note : Le terme (4,4) est 0 car eta*delta*U - mu_P = 0 à l'équilibre
# Le terme (3,3) est le terme complexe -(nu*S + mu_U + delta*P)
J = sp.Matrix([
    [-omega*nu*U - mu,  0,        -omega*nu*S,             0],
    [omega*nu*U,       -(mu+d),   omega*nu*S,              0],
    [-nu*U,            gamma,    -(nu*S + mu_U + delta*P), -delta*U],
    [0,                0,         eta*delta*P,             0]
])

print("\n--- MATRICE JACOBIENNE J(E_star) ---")
sp.pprint(J)

# ==========================================
# 3. CALCUL DES COEFFICIENTS (a1, a2, a3, a4)
# ==========================================
print("\n--- 2. CALCUL DES COEFFICIENTS DU POLYNOME ---")
char_poly = J.charpoly(lam)
coeffs = char_poly.all_coeffs()

# SymPy renvoie [1, a1, a2, a3, a4]
a1 = sp.simplify(coeffs[1])
a2 = sp.simplify(coeffs[2])
a3 = sp.simplify(coeffs[3])
a4 = sp.simplify(coeffs[4])

print("\n>>> a1 (Trace) :")
sp.pprint(a1)

print("\n>>> a4 (Déterminant) :")
sp.pprint(a4)

# ==========================================
# 4. VERIFICATION DE LA PREUVE (SIMPLIFICATIONS)
# ==========================================
print("\n--- 3. VERIFICATION ET SIMPLIFICATION ---")

# Définition de la relation à l'équilibre pour simplifier
# Relation : (mu+d)*(nu*S + mu_U + delta*P) = gamma*omega*nu*S
# On isole le terme complexe (nu*S + mu_U + delta*P)
term_complex = nu*S + mu_U + delta*P
identity_sub = (gamma * omega * nu * S) / (mu + d)

# --- A. Vérification de a4 (Déterminant) ---
# Formule manuelle : eta * delta^2 * P * U * (omega*nu*U + mu) * (mu+d)
manual_a4 = eta * delta**2 * P * U * (omega*nu*U + mu) * (mu + d)
diff_a4 = sp.simplify(a4 - manual_a4)

print(f"\n[TEST A4] Calcul SymPy vs Formule Manuelle : {'OK (Diff=0)' if diff_a4 == 0 else 'ERREUR'}")

# --- B. Vérification de Delta_4 (Le bloc 3x3 critique) ---
# C'est le mineur obtenu en supprimant ligne 4 et col 4 (indices 3,3 en python)
Delta_4 = J.minor_submatrix(3, 3).det()

print("\n[DELTA_4] Expression brute :")
sp.pprint(Delta_4)

# On applique la substitution pour voir si ça se simplifie comme prévu
# On s'attend à voir le facteur (omega*gamma - mu - d)
Delta_4_simp = sp.simplify(Delta_4.subs(term_complex, identity_sub))

print("\n[DELTA_4] Expression simplifiée avec l'identité d'équilibre :")
sp.pprint(Delta_4_simp)
# Si le terme (mu + d - gamma*omega) apparaît, c'est gagné.

# --- C. Vérification de a3 ---
# Formule preuve : a3 = - (Delta_1 + Delta_2 + Delta_3 + Delta_4)
# Calculons les 4 mineurs principaux d'ordre 3
D1 = J.minor_submatrix(0, 0).det()
D2 = J.minor_submatrix(1, 1).det()
D3 = J.minor_submatrix(2, 2).det() # Doit être 0
D4 = J.minor_submatrix(3, 3).det()

manual_a3_sum = -(D1 + D2 + D3 + D4)
diff_a3 = sp.simplify(a3 - manual_a3_sum)

print(f"\n[TEST A3] a3 (SymPy) vs Somme des Mineurs : {'OK (Diff=0)' if diff_a3 == 0 else 'ERREUR'}")

# --- D. Vérification de a2 ---
# Somme des 6 mineurs principaux d'ordre 2
sum_minors_2 = 0
import itertools
for r, c in itertools.combinations([0, 1, 2, 3], 2):
    sum_minors_2 += J[[r,c], [r,c]].det()

diff_a2 = sp.simplify(a2 - sum_minors_2)
print(f"\n[TEST A2] a2 (SymPy) vs Somme des Mineurs 2x2 : {'OK (Diff=0)' if diff_a2 == 0 else 'ERREUR'}")

print("\n--- FIN DE L'ANALYSE ---")
```

To simplify and group expressions in SymPy, you can use several functions depending on the desired outcome:

*   `sp.simplify(expr)`: This is a general-purpose simplification function that applies various algorithms to make the expression simpler. It's often a good first step, but might not always yield the *most* desired form.

*   `sp.factor(expr)`: Attempts to factor the expression into its constituent factors. This is useful if you want to see if an expression can be written as a product of simpler terms.

*   `sp.expand(expr)`: Expands products and powers in an expression. This is the opposite of `factor` and can be useful to reveal terms that might cancel out or combine.

*   `sp.collect(expr, symbol)`: Collects coefficients of a specific symbol or power of a symbol. For example, `sp.collect(a*x**2 + b*x**2 + c*x, x)` would give `(a + b)*x**2 + c*x`.

*   `sp.trigsimp(expr)`: Applies trigonometric identities to simplify expressions containing trigonometric functions.

*   `sp.nsimplify(expr)`: Converts a float to a rational expression, which can sometimes lead to simpler forms if the floats are approximations of rational numbers.

*   `sp.powsimp(expr)`: Simplifies expressions containing powers.

*   `sp.radsimp(expr)`: Rationalizes the denominator of an expression.

When working with complex expressions, you might need to apply these functions iteratively or in combination to reach the desired level of 'simplicity'.
